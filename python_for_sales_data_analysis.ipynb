## Menjalankan SQL di Colab dengan `sqlite3`

```python
#17
#Menjalankan SQL di Colab
from sqlite3 import connect
conn = connect(':memory:')
df.to_sql('tokopaedi',conn, index=False, if_exists='replace')
```

```python
#18
#Query SQL
df_sql = pd.read_sql("""
SELECT
    tokopaedi.city,
    sum(tokopaedi.profit) profit
from
    tokopaedi
group by
    1
order by
    2 desc
limit 10
""", conn)
df_sql
```

Tampilkan TOP 5 City selama 2016 dengan berdasarkan nilai profit yang paling tinggi. Tampilkan juga dengan bar diagram.

```python
# Filter data untuk tahun 2016.
df_2016 = df[df['year'] == 2016]

# Kelompokkan data berdasarkan kota dan hitung total keuntungan untuk setiap kota.
df_city_profit = df_2016.groupby('city')['profit'].sum().reset_index()

# Urutkan data berdasarkan keuntungan dalam urutan menurun dan pilih 5 kota teratas.
top_5_cities = df_city_profit.sort_values('profit', ascending=False).head(5)
top_5_cities
```

```python
# Buat diagram batang untuk memvisualisasikan 5 kota teratas dan keuntungannya.
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.bar(top_5_cities['city'], top_5_cities['profit'])
plt.xlabel('City')
plt.ylabel('Profit')
plt.title('Top 5 Cities by Highest Profit in 2016')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```
